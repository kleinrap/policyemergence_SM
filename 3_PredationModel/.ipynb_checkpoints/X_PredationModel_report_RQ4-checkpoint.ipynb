```python
'''
This notebook is used to answer research question 4:
What impact the electorate has (between scenarios 0, 4 and 5) on the policy process
emergence (benchmark scenario) model.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import ast
import copy
from copy import deepcopy
```

```python
rep_runs = 50 # repetitions
sce = 7

linestyles = ['-',':', '--','-', ':', '--','-', '--', ':','-', '--', ':','-', '--', ':','-', '--', ':']
```

```python
'''
Analysis of the predation model - WITH POLICY EMERGENCE
Main parameters of interest (the rest is constant):
- Wolf count
- Sheep count
- Grass patches count
'''

steps_t = []
sheep_t = []
wolves_t = []
grass_t = []
sheepR_t = []
wolvesR_t = []
grassR_t = []


# going through all of the scenarios
for sce_i in range(sce):
    
    predation_model = []

    steps = []
    sheep = []
    wolves = []
    grass = []
    sheepR = []
    wolvesR = []
    grassR = []

    # We first read all the files into arrays that contains each entire file
    for i in range(rep_runs):
        if sce_i == 0:
            predation_model.append(pd.read_csv('O_Pre_model_Sce0_Run' + str(i) + '_el0.0.csv'))
        if sce_i == 1:
            predation_model.append(pd.read_csv('O_Pre_model_Sce4_Run' + str(i) + '_el0.02.csv'))
        if sce_i == 2:
            predation_model.append(pd.read_csv('O_Pre_model_Sce4_Run' + str(i) + '_el0.2.csv'))
        if sce_i == 3:
            predation_model.append(pd.read_csv('O_Pre_model_Sce4_Run' + str(i) + '_el0.5.csv'))
        if sce_i == 4:
            predation_model.append(pd.read_csv('O_Pre_model_Sce5_Run' + str(i) + '_el0.02.csv'))
        if sce_i == 5:
            predation_model.append(pd.read_csv('O_Pre_model_Sce5_Run' + str(i) + '_el0.2.csv'))
        if sce_i == 6:
            predation_model.append(pd.read_csv('O_Pre_model_Sce5_Run' + str(i) + '_el0.5.csv'))
        
        steps.append([])
        sheep.append([])
        wolves.append([])
        grass.append([])
        sheepR.append([])
        wolvesR.append([])
        grassR.append([])
    
    print('Scenario', str(sce_i), 'files opened')
        
    print(predation_model[0].head())
    print(list(predation_model[0].head(0)))

    # saving the paremeters in different arrays
    for i in range(rep_runs):
        for index, row in predation_model[i].iterrows():
            steps[i].append(index)
            sheep[i].append(row['Sheep'])
            wolves[i].append(row['Wolves'])
            grass[i].append(row['Grass'])
            sheepR[i].append(row['SheepRepro'])
            wolvesR[i].append(row['WolfRepro'])
            grassR[i].append(row['GrassRegrowth'])

    # saving all scenarios into arrays
    steps_t.append(copy.deepcopy(steps))
    sheep_t.append(copy.deepcopy(sheep))
    wolves_t.append(wolves)
    grass_t.append(grass)
    sheepR_t.append(sheepR)
    wolvesR_t.append(wolvesR)
    grassR_t.append(grassR)

print('All files saved')
```

```python
'''
Getting the Kernel Density Estimates for all parameters saved
WITH PE results
'''

sheep_t_KDE_100 = []
wolves_t_KDE_100 = []
grass_t_KDE_100 = []

sheep_t_KDE_600 = []
wolves_t_KDE_600 = []
grass_t_KDE_600 = []

sheep_t_KDE_1100 = []
wolves_t_KDE_1100 = []
grass_t_KDE_1100 = []

sheep_t_KDE_1600 = []
wolves_t_KDE_1600 = []
grass_t_KDE_1600 = []
    
for sce_i in range(sce):
    
    sheep_t_KDE_100.append([])
    wolves_t_KDE_100.append([])
    grass_t_KDE_100.append([])

    sheep_t_KDE_600.append([])
    wolves_t_KDE_600.append([])
    grass_t_KDE_600.append([])

    sheep_t_KDE_1100.append([])
    wolves_t_KDE_1100.append([])
    grass_t_KDE_1100.append([])

    sheep_t_KDE_1600.append([])
    wolves_t_KDE_1600.append([])
    grass_t_KDE_1600.append([])

    # saving/sorting
    for i in range(rep_runs):
        sheep_t_KDE_100[sce_i].append(round(sheep_t[sce_i][i][99],4))
        wolves_t_KDE_100[sce_i].append(round(wolves_t[sce_i][i][99],4))
        grass_t_KDE_100[sce_i].append(round(grass_t[sce_i][i][99],4))

        sheep_t_KDE_600[sce_i].append(round(sheep_t[sce_i][i][599],4))
        wolves_t_KDE_600[sce_i].append(round(wolves_t[sce_i][i][599],4))
        grass_t_KDE_600[sce_i].append(round(grass_t[sce_i][i][599],4))

        sheep_t_KDE_1100[sce_i].append(round(sheep_t[sce_i][i][1099],4))
        wolves_t_KDE_1100[sce_i].append(round(wolves_t[sce_i][i][1099],4))
        grass_t_KDE_1100[sce_i].append(round(grass_t[sce_i][i][1099],4))

        sheep_t_KDE_1600[sce_i].append(round(sheep_t[sce_i][i][-101],4))
        wolves_t_KDE_1600[sce_i].append(round(wolves_t[sce_i][i][-101],4))
        grass_t_KDE_1600[sce_i].append(round(grass_t[sce_i][i][-101],4)) 

    # pdf = kde_funcs(x, evenness_KDE_a, bandwidth=0.2)
    hist = plt.hist(sheep_t_KDE_1600[sce_i], bins=10, density=True, alpha = 0.5, orientation="horizontal")
    hist2 = plt.hist(wolves_t_KDE_1600[sce_i], bins=10, density=True, alpha = 0.5, orientation="horizontal")
```
