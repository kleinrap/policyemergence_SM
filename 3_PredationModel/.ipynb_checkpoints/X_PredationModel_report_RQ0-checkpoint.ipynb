```python
'''
This notebook is used to answer research question 0:
What is the difference between the predation alone model and the predation with policy
emergence (benchmark scenario) model.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import ast
import copy
from copy import deepcopy
```

```python
rep_runs = 50 # repetitions

linestyles = ['-',':', '--','-', ':', '--','-', '--', ':','-', '--', ':','-', '--', ':','-', '--', ':']
```

```python
'''
Analysis of the predation model - ALONE
Main parameters of interest (the rest is constant):
- Wolf count
- Sheep count
- Grass patches count
'''

predation_model = []

# the different columns of the panda - ['Sheep', 'Wolves', 'Grass']
steps_a = []
sheep_a = []
wolves_a = []
grass_a = []
sheepR_a = []
wolvesR_a = []
grassR_a = []

# We first read all the files into arrays that contains each entire file
for i in range(rep_runs):

    predation_model.append(pd.read_csv('O_Pre_alone_model_Run' + str(i) + '.csv'))

    steps_a.append([])
    sheep_a.append([])
    wolves_a.append([])
    grass_a.append([])
    sheepR_a.append([])
    wolvesR_a.append([])
    grassR_a.append([])

print('Alone files opened')

# print(list(predation_model[0].head(0))) # reading the head of the panda

# saving the paremeters in different arrays
for i in range(rep_runs):
    for index, row in predation_model[i].iterrows():
        steps_a[i].append(index)
        sheep_a[i].append(row['Sheep'])
        wolves_a[i].append(row['Wolves'])
        grass_a[i].append(row['Grass'])
        sheepR_a[i].append(row['SheepRepro'])
        wolvesR_a[i].append(row['WolfRepro'])
        grassR_a[i].append(row['GrassRegrowth'])
        
print('Alone files saved')
```

```python
'''
Analysis of the predation model - WITH POLICY EMERGENCE
Main parameters of interest (the rest is constant):
- Wolf count
- Sheep count
- Grass patches count
'''

# steps_t = []
# sheep_t = []
# wolves_t = []
# grass_t = []
# sheepR_t = []
# wolvesR_t = []
# grassR_t = []

predation_model = []

steps = []
sheep = []
wolves = []
grass = []
sheepR = []
wolvesR = []
grassR = []

# We first read all the files into arrays that contains each entire file
for i in range(rep_runs):

    predation_model.append(pd.read_csv('O_Pre_model_Sce0_Run' + str(i) + '_el0.0.csv'))

    steps.append([])
    sheep.append([])
    wolves.append([])
    grass.append([])
    sheepR.append([])
    wolvesR.append([])
    grassR.append([])

print('Files opened')

print(predation_model[0].head())
print(list(predation_model[0].head(0)))

# saving the paremeters in different arrays
for i in range(rep_runs):
    for index, row in predation_model[i].iterrows():
        steps[i].append(index)
        sheep[i].append(row['Sheep'])
        wolves[i].append(row['Wolves'])
        grass[i].append(row['Grass'])
        sheepR[i].append(row['SheepRepro'])
        wolvesR[i].append(row['WolfRepro'])
        grassR[i].append(row['GrassRegrowth'])

# # saving all scenarios into arrays
# steps_t.append(copy.deepcopy(steps))
# sheep_t.append(copy.deepcopy(sheep))
# wolves_t.append(wolves)
# grass_t.append(grass)
# sheepR_t.append(sheepR)
# wolvesR_t.append(wolvesR)
# grassR_t.append(grassR)

print('All files saved')
```

```python
'''
Getting the Kernel Density Estimates for all parameters saved
ALONE results
'''


sheep_a_KDE_100 = []
wolves_a_KDE_100 = []
grass_a_KDE_100 = []

sheep_a_KDE_600 = []
wolves_a_KDE_600 = []
grass_a_KDE_600 = []

sheep_a_KDE_1100 = []
wolves_a_KDE_1100 = []
grass_a_KDE_1100 = []

sheep_a_KDE_1600 = []
wolves_a_KDE_1600 = []
grass_a_KDE_1600 = []

# saving/sorting
for i in range(rep_runs):
    sheep_a_KDE_100.append(round(sheep_a[i][99],4))
    wolves_a_KDE_100.append(round(wolves_a[i][99],4))
    grass_a_KDE_100.append(round(grass_a[i][99],4))
    
    sheep_a_KDE_600.append(round(sheep_a[i][599],4))
    wolves_a_KDE_600.append(round(wolves_a[i][599],4))
    grass_a_KDE_600.append(round(grass_a[i][599],4))
    
    sheep_a_KDE_1100.append(round(sheep_a[i][1099],4))
    wolves_a_KDE_1100.append(round(wolves_a[i][1099],4))
    grass_a_KDE_1100.append(round(grass_a[i][1099],4))
    
    sheep_a_KDE_1600.append(round(sheep_a[i][-1],4))
    wolves_a_KDE_1600.append(round(wolves_a[i][-1],4))
    grass_a_KDE_1600.append(round(grass_a[i][-1],4)) 

# pdf = kde_funcs(x, evenness_KDE_a, bandwidth=0.2)
hist = plt.hist(sheep_a_KDE_1600, bins=10, density=True, alpha = 0.5, orientation="horizontal")
hist2 = plt.hist(wolves_a_KDE_1600, bins=10, density=True, alpha = 0.5, orientation="horizontal")
```

```python
'''
Getting the Kernel Density Estimates for all parameters saved
WITH PE results
'''

sheep_t_KDE_100 = []
wolves_t_KDE_100 = []
grass_t_KDE_100 = []

sheep_t_KDE_600 = []
wolves_t_KDE_600 = []
grass_t_KDE_600 = []

sheep_t_KDE_1100 = []
wolves_t_KDE_1100 = []
grass_t_KDE_1100 = []

sheep_t_KDE_1600 = []
wolves_t_KDE_1600 = []
grass_t_KDE_1600 = []
    
# for w_el_i in range(len(w_el)):
    
#     sheep_t_KDE_100.append([])
#     wolves_t_KDE_100.append([])
#     grass_t_KDE_100.append([])

#     sheep_t_KDE_600.append([])
#     wolves_t_KDE_600.append([])
#     grass_t_KDE_600.append([])

#     sheep_t_KDE_1100.append([])
#     wolves_t_KDE_1100.append([])
#     grass_t_KDE_1100.append([])

#     sheep_t_KDE_1600.append([])
#     wolves_t_KDE_1600.append([])
#     grass_t_KDE_1600.append([])

# saving/sorting
for i in range(rep_runs):
    sheep_t_KDE_100.append(round(sheep[i][99],4))
    wolves_t_KDE_100.append(round(wolves[i][99],4))
    grass_t_KDE_100.append(round(grass[i][99],4))

    sheep_t_KDE_600.append(round(sheep[i][599],4))
    wolves_t_KDE_600.append(round(wolves[i][599],4))
    grass_t_KDE_600.append(round(grass[i][599],4))

    sheep_t_KDE_1100.append(round(sheep[i][1099],4))
    wolves_t_KDE_1100.append(round(wolves[i][1099],4))
    grass_t_KDE_1100.append(round(grass[i][1099],4))

    sheep_t_KDE_1600.append(round(sheep[i][-101],4))
    wolves_t_KDE_1600.append(round(wolves[i][-101],4))
    grass_t_KDE_1600.append(round(grass[i][-101],4)) 

# pdf = kde_funcs(x, evenness_KDE_a, bandwidth=0.2)
hist = plt.hist(sheep_t_KDE_1600, bins=10, density=True, alpha = 0.5, orientation="horizontal")
hist2 = plt.hist(wolves_t_KDE_1600, bins=10, density=True, alpha = 0.5, orientation="horizontal")
```

```python
'''
The counts - PLOTTING
SHEEP AND WOLF PER SCENARIO WITH PE
'''

fig = plt.figure(tight_layout=True, figsize=(15,12))
# 3 height (per graph) and 15 width
gs = gridspec.GridSpec(4, 8)


'''SHEEP WOLF WITHOUT PE'''
# number of agents (sheep)
axarr2 = fig.add_subplot(gs[0, 0:5])
axarr2.plot(steps_a[0], sheep_a[0], color = 'b', label='Sheep')
axarr2.plot(steps_a[0], wolves_a[0], color = 'r', label='Wolves')
for i in range(rep_runs):
    axarr2.plot(steps_a[i], sheep_a[i], color = 'b')
    axarr2.plot(steps_a[i], wolves_a[i], color = 'r')
axarr2.set_title('Sheep/Wolves (without policy emergence)')
axarr2.set_xlim([0, 1600])
axarr2.set_ylim([0, 350])
axarr2.set_ylabel('# of agents')
axarr2.legend
axarr2.legend(loc="upper left")
axarr2.grid(True)

# KDE (sheep/wolf) - 600
axarr2 = fig.add_subplot(gs[0, 5:6])
axarr2.hist(sheep_a_KDE_600, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_a_KDE_600, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 600')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)

# KDE (sheep/wolf) - 1100
axarr2 = fig.add_subplot(gs[0, 6:7])
axarr2.hist(sheep_a_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_a_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 1100')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)

# KDE (sheep/wolf) - 1600
axarr2 = fig.add_subplot(gs[0, 7:8])
axarr2.hist(sheep_a_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_a_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 1600')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)


'''SHEEP WOLF WITH PE SCENARIO 0'''
# number of agents (sheep)
axarr2 = fig.add_subplot(gs[1, 0:5])
axarr2.plot(steps[0], sheep[0], color = 'b', label='Sheep')
axarr2.plot(steps[0], wolves[0], color = 'r', label='Wolves')
for i in range(rep_runs):
    axarr2.plot(steps[i], sheep[i], color = 'b')
    axarr2.plot(steps[i], wolves[i], color = 'r')
axarr2.set_title('Sheep/Wolves (Scenario Benchmark)')
axarr2.set_xlim([0, 1600])
axarr2.set_ylim([0, 350])
axarr2.set_ylabel('# of agents')
axarr2.legend
axarr2.legend(loc="upper left")
axarr2.grid(True)

# KDE (sheep/wolf) - 600
axarr2 = fig.add_subplot(gs[1, 5:6])
axarr2.hist(sheep_t_KDE_600, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_t_KDE_600, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 600')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)

# KDE (sheep/wolf) - 1100
axarr2 = fig.add_subplot(gs[1, 6:7])
axarr2.hist(sheep_t_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_t_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 1100')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)

# KDE (sheep/wolf) - 1600
axarr2 = fig.add_subplot(gs[1, 7:8])
axarr2.hist(sheep_t_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'b')
axarr2.hist(wolves_t_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'r')
axarr2.set_title('t = 1600')
axarr2.set_xlim([0, 0.050])
axarr2.set_ylim([0, 350])
axarr2.grid(True)

'''SHEEP WOLF WITHOUT PE'''
# number of agents (grass)
axarr2 = fig.add_subplot(gs[2, 0:5])
axarr2.plot(steps_a[0], grass_a[0], color = 'g', label='Grass')
for i in range(rep_runs):
    axarr2.plot(steps_a[i], grass_a[i], color = 'g')
axarr2.set_title('Grass (without policy emergence)')
axarr2.set_xlim([0, 1600])
axarr2.set_ylim([500, 2200])
axarr2.set_ylabel('# of patches')
axarr2.legend
axarr2.legend(loc="upper left")
axarr2.grid(True)

# KDE (grass) - 600
axarr2 = fig.add_subplot(gs[2, 5:6])
axarr2.hist(grass_a_KDE_600, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 600')
axarr2.set_xlim([0, 0.005])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

# KDE (grass) - 1100
axarr2 = fig.add_subplot(gs[2, 6:7])
axarr2.hist(grass_a_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 1100')
axarr2.set_xlim([0, 0.005])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

# KDE (grass) - 1600
axarr2 = fig.add_subplot(gs[2, 7:8])
axarr2.hist(grass_a_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 1600')
axarr2.set_xlim([0, 0.0050])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

'''SHEEP WOLF WITH PE SCENARIO 0'''
# number of agents (grass)
axarr2 = fig.add_subplot(gs[3, 0:5])
axarr2.plot(steps[0], grass[0], color = 'g', label='Grass')
for i in range(rep_runs):
    axarr2.plot(steps[i], grass[i], color = 'g')
axarr2.set_title('Grass (Scenario Benchmark)')
axarr2.set_xlim([0, 1600])
axarr2.set_ylim([500, 2200])
axarr2.set_ylabel('# of patches')
axarr2.legend
axarr2.legend(loc="upper left")
axarr2.grid(True)

# KDE (grass) - 600
axarr2 = fig.add_subplot(gs[3, 5:6])
axarr2.hist(grass_t_KDE_600, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 600')
axarr2.set_xlim([0, 0.005])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

# KDE (grass) - 1100
axarr2 = fig.add_subplot(gs[3, 6:7])
axarr2.hist(grass_t_KDE_1100, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 1100')
axarr2.set_xlim([0, 0.005])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

# KDE (grass) - 1600
axarr2 = fig.add_subplot(gs[3, 7:8])
axarr2.hist(grass_t_KDE_1600, bins=10, density=True, orientation="horizontal", color = 'g')
axarr2.set_title('t = 1600')
axarr2.set_xlim([0, 0.0050])
axarr2.set_ylim([500, 2200])
axarr2.grid(True)

fig.savefig('Predation_Q0_pops.eps', format='eps', dpi=1000)
```
